```python
import ast

# Extract HairsSpecializesIn values
hair_specs = philly_salons['attributes'].dropna().apply(
    lambda x: ast.literal_eval(str(x)).get('HairSpecializesIn')
    if isinstance(x, dict) else None
)
print(hair_specs.value_counts().head(20))
```

```python
# Clean and extract HairSpecializesIn properly
def get_hair_spec(attr):
    try:
        if isinstance(attr, dict):
            return attr.get('HairSpecializesIn')
        elif isinstance(attr, str):
            return ast.literal_eval(attr).get('HairSpecializesIn')
    except:
        return None

philly_salons['hair_specialization'] = philly_salons['attributes'].apply(get_hair_spec)

# See clean counts
print(philly_salons['hair_specialization'].value_counts())
print(f"\nMissing values: {philly_salons['hair_specialization'].isna().sum()}")
```

```python
salon_ratings = pd.read_csv('/content/drive/MyDrive/philly salon ratings - Sheet1 (2).csv')
print(f"Salons loaded: {len(salon_ratings)}")
print(salon_ratings.columns.tolist())
```

```python
# Clean new columns
salon_ratings['market_focus'] = salon_ratings['market_focus'].str.strip().str.title()
salon_ratings['secondary_focus'] = salon_ratings['secondary_focus'].str.strip().str.title()

# Check market focus breakdown
print("Market Focus Categories:")
print(salon_ratings['market_focus'].value_counts())
```

```python
# Analyze by market focus
market_analysis = salon_ratings.groupby('market_focus').agg(
    salon_count=('name', 'count'),
    avg_stars=('stars', 'mean'),
    avg_reviews=('review_count', 'mean'),
    avg_min_price=('price_min', 'mean'),
    avg_max_price=('price_max', 'mean')
).round(2)

print(market_analysis.to_string())
```

```python
# Calculate price-to-rating efficiency
salon_ratings['avg_price'] = (salon_ratings['price_min'] + salon_ratings['price_max']) / 2
salon_ratings['stars_per_dollar'] = (salon_ratings['stars'] / salon_ratings['avg_price'] * 100).round(3)
salon_ratings['reviews_per_dollar'] = (salon_ratings['review_count'] / salon_ratings['avg_price']).round(2)

# Group by market focus
efficiency_analysis = salon_ratings.groupby('market_focus').agg(
    avg_price=('avg_price', 'mean'),
    avg_stars_per_dollar=('stars_per_dollar', 'mean'),
    avg_reviews_per_dollar=('reviews_per_dollar', 'mean')
).round(2)

print(efficiency_analysis.to_string())
```

```python
# Analyze by zip code
zip_analysis = salon_ratings.groupby('postal_code').agg(
    salon_count=('name', 'count'),
    avg_stars=('stars', 'mean'),
    avg_reviews=('review_count', 'mean'),
    avg_min_price=('price_min', 'mean'),
    avg_max_price=('price_max', 'mean')
).round(2)

print(zip_analysis.sort_values('salon_count', ascending=False).head(10))
```

```python
# Save zip code analysis
zip_analysis = salon_ratings.groupby('postal_code').agg(
    salon_count=('name', 'count'),
    avg_stars=('stars', 'mean'),
    avg_reviews=('review_count', 'mean'),
    avg_min_price=('price_min', 'mean'),
    avg_max_price=('price_max', 'mean')
).round(2).sort_values('avg_stars', ascending=False)

zip_analysis.to_csv('/content/drive/MyDrive/zip_analysis_results.csv', index=True)
print("Saved!")

# Also save the full cleaned salon dataset with all our new columns
salon_ratings.to_csv('/content/drive/MyDrive/philly_salons_FINAL.csv', index=False)
print("Final dataset saved!")
```
